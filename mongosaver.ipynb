```python
from AgentNet.Agent.langgraph_utils import MongoDBSaver
from AgentNet.config import nosql_service
```

```python
memory_manager = MongoDBSaver(client = nosql_service, db_name = "chat_history")
```

```python
config =  {"configurable": {"thread_id": "DFT_Agent556ddd12-2322-4d73-99e9-73da6e387bbc", "recursion_limit": 100}}
```

```python
memory_manager.get(config)
```

```python
#delete the last 24 data in nosql
collection = nosql_service["chat_history"]["checkpoints"]
```

```python
# Define the filter
thread_id_filter = {"thread_id": "DFT_Agent556ddd12-2322-4d73-99e9-73da6e387bbc"}

# Fetch the last 24 messages based on natural order (reverse of insertion order)
last_24_messages = (
    collection.find(thread_id_filter)
    .sort("_id", -1)  # Sort by _id in descending order (latest first)
    .limit(10)
)

# Extract their IDs
ids_to_delete = [message["_id"] for message in last_24_messages]
delete_result = collection.delete_many({"_id": {"$in": ids_to_delete}})
```

```python
# Define the filter to match documents where thread_id does NOT contain "DFT_Agent"
filter_query = {"thread_id": {"$not": {"$regex": "DFT_Agent"}}}

# Delete the matching documents
delete_result = collection.delete_many(filter_query)

print(f"Deleted {delete_result.deleted_count} documents.")
```
